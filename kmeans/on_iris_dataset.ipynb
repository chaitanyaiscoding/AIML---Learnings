# Iris is Dataset gives Data on 3 different types of flowers


```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data)
X.columns = iris.feature_names
y = iris.target

X
# y
```

```python
# Dimensionality Reduction

X = iris.data
y = iris.target

# step-1 - visualizing

# sns.scatterplot(X)

sns.scatterplot(x=X[:,0],y=X[:,2],c=y)
```

```python
#Scaling the Data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_Scaled = scaler.fit_transform(X)

X_Scaled
```

```python
#finding exact K value from WCSS

wcss = []
for k in range (1,21):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_Scaled)
    wcss.append(kmeans.inertia_)

wcss 


sns.lineplot(x=range(1,21),y=wcss,marker='o')
```

```python
#KMeans

kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(X_Scaled)

sns.scatterplot(x=X_Scaled[:,0],y=X_Scaled[:,2],c=labels)
```

```python
#predicting on new data

data_new = scaler.transform([[5.1, 3.5, 1.4, 0.2]])
new_pred = kmeans.predict(data_new)

print(new_pred)
```
