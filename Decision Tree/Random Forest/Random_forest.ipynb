```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
Dataset = pd.read_csv("shop_smart_ecommerce.csv")
Dataset.head()

X = Dataset.drop(columns=["Revenue"])
y = Dataset["Revenue"].astype(int)

X.head()
y.head()

le = LabelEncoder()
X["Month"] = le.fit_transform(X["Month"])
X["VisitorType"] = le.fit_transform(X["VisitorType"])

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred_test = model.predict(X_test)
y_pred_training = model.predict(X_train)

print("Training Accuracy",accuracy_score(y_train,y_pred_training * 100),"%")
print("Testing Accuracy",accuracy_score(y_test,y_pred_test * 100),"%")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=201,oob_score=True)

rf.fit(X_train,y_train)

y_pred = rf.predict(X_test)

print("OOB score",rf.oob_score_,"%")
print("Testing Accuracy",accuracy_score(y_test,y_pred * 100),"%")
```
